# 画像の読み込み


ローカル画像を読み込むには

```c
cv.imread('画像のPath', 読み込み形式)
```

を使います  
読み込み形式は

- cv.IMREAD_COLOR(BGR)(デフォルト)
- cv.IMREAD_UNCHANGED(BGR、アルファチャンネルが含まれる場合は BGRA)
- cv.IMREAD_GRAYSCALE(グレースケール)

を指定します

[公式チュートリアル](https://docs.opencv.org/4.x/db/deb/tutorial_display_image.html)


## 試しに読み込み(256 x 256 の jpg と png)


### jpg(アルファチャンネルなし)を読み込んでみる


````python
tuples = [('IMREAD_COLOR', cv.IMREAD_COLOR),
          ('IMREAD_UNCHANGED', cv.IMREAD_UNCHANGED),
          ('IMREAD_GRAYSCALE', cv.IMREAD_GRAYSCALE)]

for tuple in tuples:
    print(tuple[0])
    img = cv.imread('pochi_256_256.jpg', tuple[1])
    print(f'type:{type(img)} ndim:{img.ndim} shape:{img.shape}\n')
````

### png(アルファチャンネルあり)を読み込んでみる


````python
tuples = [('IMREAD_COLOR', cv.IMREAD_COLOR),
          ('IMREAD_UNCHANGED', cv.IMREAD_UNCHANGED),
          ('IMREAD_GRAYSCALE', cv.IMREAD_GRAYSCALE)]

for tuple in tuples:
    print(tuple[0])
    img = cv.imread('pochi_256_256.png', tuple[1])
    print(f'type:{type(img)} ndim:{img.ndim} shape:{img.shape}\n')
````

## つまり？


読み込んだ画像は画像種別問わず **[高さ, 幅, 種別毎のパラメータ]** の 2(グレースケール) or 3 次元(BGR or BGRA)配列(ndarray)になります

- グレースケール  
  mat[y, x]で読み書き

- 3 次元  
  mat[y, x, BGR(A)]で読み書き

# notebook へ画像を表示


ここでは、matplotlib を用いて表示方法を説明します


## 単純な画像表示


````python
img = cv.imread('pochi_256_256.png', cv.IMREAD_COLOR)
print(f'shape:{img.shape}')

# 画像表示
plt.figure(figsize=(8, 16))
plt.imshow(img)
plt.show()
````

元画像と見比べると分かる通り、色がおかしいです  
これは、imshow 関数では入力画像の色空間を RGB(A)と解釈するのに対し、  
OpenCv で読み込んだ画像の色空間が BGR(A)であるためです  
ここでは、画像の色空間の変換方法を説明します


## 色空間を変換して表示


````python
img = cv.imread('pochi_256_256.png', cv.IMREAD_COLOR)
print(f'shape:{img.shape}')

# BGRをRGBに変換
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# 画像表示
plt.figure(figsize=(8, 16))
plt.imshow(img2)
plt.show()
````

色空間の変換には

```c
cv.cvtColor(画像, 色空間変換コード)
```

を使用します

ここでは **BGR から RGB** に変換しましたが、
他にもグレースケールへの変換や  
アルファチャンネルの追加・削除等、多彩な変換が行えます

色空間変換コードの種類は[公式の ColorConversionCodes](https://docs.opencv.org/4.x/d8/d01/group__imgproc__color__conversions.html#ga4e0972be5de079fed4e3a10e24ef5ef0)で確認出来ます


## グレースケール画像の表示


````python
img = cv.imread('pochi_256_256.png', cv.IMREAD_GRAYSCALE)
print(f'shape:{img.shape}')

# 画像表示
plt.figure(figsize=(8, 16))
plt.imshow(img)
plt.show()
````

グレースケールを表示させたいのに色がおかしくなりました  
これは imshow 関数で扱うカラーマップが、デフォルトでは"viridis"というカラーマップを用いる為です  
グレースケール用のカラーマップを指定するには引数 cmap に"gray"を指定します  
また、OpenCv ではグレースケール値の上下限が 0-255 の範囲なので、vmin、vmax に指定しておきます


````python
img = cv.imread('pochi_256_256.png', cv.IMREAD_GRAYSCALE)
print(f'shape:{img.shape}')

# 画像表示
plt.figure(figsize=(8, 16))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
````

## 画像表示用関数を作成する


ここまで画像表示方法を見てきましたが、表示の度に色空間に応じて変換をかけるのは正直面倒です  
また、同時に複数枚表示したいケースもある為、画像表示用関数を作成します  
※エラーチェックなし、機能も必要最低限です


````python
def imgshow(src):
    # 複数枚
    if type(src) is list:
        fig = plt.figure(figsize=(8 * len(src), 16))
        for i, img in enumerate(src, 1):
            ax = fig.add_subplot(1, len(src), i)
            # グレースケール
            if img.ndim == 2:
                ax.imshow(img, cmap='gray', vmin=0, vmax=255)
            # BGR
            elif img.shape[2] == 3:
                ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
            # BGRA
            elif img.shape[2] == 4:
                ax.imshow(cv.cvtColor(img, cv.COLOR_BGRA2RGBA))
    # 1枚
    else:
        plt.figure(figsize=(8, 16))
        if src.ndim == 2:
            plt.imshow(src, cmap='gray', vmin=0, vmax=255)
        elif src.shape[2] == 3:
            plt.imshow(cv.cvtColor(src, cv.COLOR_BGR2RGB))
        elif src.shape[2] == 4:
            plt.imshow(cv.cvtColor(src, cv.COLOR_BGRA2RGBA))

    plt.show()
````

### 画像表示用関数を使用してみる


#### BGR


````python
img = cv.imread('pochi_256_256.png', cv.IMREAD_COLOR)
print(f'shape:{img.shape}')

imgshow(img)
````

#### BGRA


````python
img = cv.imread('pochi_256_256.png', cv.IMREAD_UNCHANGED)
print(f'shape:{img.shape}')

imgshow(img)
````

#### グレースケール


````python
img = cv.imread('pochi_256_256.png', cv.IMREAD_GRAYSCALE)
print(f'shape:{img.shape}')

imgshow(img)
````

#### 上記三種を同時表示


````python
images = []
images.append(cv.imread('pochi_256_256.png', cv.IMREAD_COLOR))
images.append(cv.imread('pochi_256_256.png', cv.IMREAD_UNCHANGED))
images.append(cv.imread('pochi_256_256.png', cv.IMREAD_GRAYSCALE))
imgshow(images)
````
